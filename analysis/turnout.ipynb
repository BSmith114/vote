```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reads csvs into dataframes
cvap = pd.read_csv('../data/cvap.csv', index_col=['fips', 'election'])
vote = pd.read_csv('../data/vote.csv', index_col=['fips', 'election'])

# gets vote totals from vote
vote_total = vote[['state', 'county', 'democrat', 'republican', 'other', 'total', 'democrat_margin_percent']]
vote_total = vote_total.rename(columns={'democrat_margin_percent': 'dem_margin'})

# creates new df with turnout data
df = cvap.join(vote_total)
df['turnout'] = df.total / df.cvap

columns = ['state', 'county', 'turnout', 'cvap', 'dem_margin']

df[columns][df['state'] == 'Florida']
```
