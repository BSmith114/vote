```python
# gets vote totals from vote 
vote_total = vote.vote[['fips', 'state', 'county', 'election', 'democrat', 'republican', 'other', 'total', 'democrat_margin_percent']]
vote_total = vote_total.rename(columns={'democrat_margin_percent': 'dem_margin'})

# creates new df with turnout data
df = pd.merge(cvap, vote_total, on=['fips', 'election'])
df['turnout'] = df.total / df.cvap
df.sort_values(['fips', 'election'], inplace=True)
df.reset_index(inplace=True)
df.drop(['index', 'level_0'], axis=1, inplace=True)

# adds turnout columns
df = df.join(df[['fips', 'dem_margin' ,'turnout']].groupby('fips').diff(), rsuffix='_change')

# filters to just 2016 election
df.query('election == 2016', inplace=True)

# resets index
df.reset_index(inplace=True)
df.drop(['index'], axis=1, inplace=True)

columns = ['state', 'county', 'turnout', 'cvap', 'dem_margin', 'turnout_change']

df[columns][df['state'] == 'Florida'].sort_values(['turnout_change'])
```
